```
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

from google.colab import drive
drive.mount('/content/drive')
```

```
data0 = pd.read_csv('/content/drive/MyDrive/Topic/Final_train/datav3/none_v3.csv')
data1 = pd.read_csv('/content/drive/MyDrive/AI/Project/datav3/blocking_v3.csv')
data2 = pd.read_csv('/content/drive/MyDrive/AI/Project/datav3/holding_v3.csv')
data3 = pd.read_csv('/content/drive/MyDrive/AI/Project/datav3/pushing_v3.csv')
data4 = pd.read_csv('/content/drive/MyDrive/AI/Project/datav3/newoff_v3.csv')
```

```
data0.loc[:,'none'] = 1
data0.loc[:,'blocking'] = 0
data0.loc[:,'holding'] = 0
data0.loc[:,'pushing'] = 0
data0.loc[:,'offensive foul'] = 0
data0 = data0.iloc[:,1:]

data1.loc[:,'none'] = 0
data1.loc[:,'blocking'] = 1
data1.loc[:,'holding'] = 0
data1.loc[:,'pushing'] = 0
data1.loc[:,'offensive foul'] = 0
data1 = data1.iloc[:,1:]

data2.loc[:,'none'] = 0
data2.loc[:,'blocking'] = 0
data2.loc[:,'holding'] = 1
data2.loc[:,'pushing'] = 0
data2.loc[:,'offensive foul'] = 0
data2 = data2.iloc[:,1:]

data3.loc[:,'none'] = 0
data3.loc[:,'blocking'] = 0
data3.loc[:,'holding'] = 0
data3.loc[:,'pushing'] = 1
data3.loc[:,'offensive foul'] = 0
data3 = data3.iloc[:,1:]

data4.loc[:,'none'] = 0
data4.loc[:,'blocking'] = 0
data4.loc[:,'holding'] = 0
data4.loc[:,'pushing'] = 0
data4.loc[:,'offensive foul'] = 1
data4 = data4.iloc[:,1:]

data = pd.concat([data0,data1,data2,data3,data4],ignore_index=True)
data.head()
```

```
data.info()
```

```
X_train = data.iloc[:,:24].copy()

Y_train = data.iloc[:,24].copy()
```

```
X_train, X_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.25, random_state=1, stratify=Y_train)

print(X_train.shape)
print(y_train.shape)
```

```
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=50,p=2.3,metric='minkowski')
knn.fit(X_train, y_train)


result = knn.predict(X_test)
# result = np.argmax(knn.predict(X_test),axis=1)
# old_y_test = np.argmax(y_test,axis=1)

test_acc= len(result[result == y_test])/len(y_test) * 100

print(result[:10])
print(test_acc)
# print(len(result[result == y_test]))
# print(len(y_test))
# print(f"Test accuracy:{test_acc}%")
```

```
result = knn.predict_proba(X_test)
print(result[:10])
```
